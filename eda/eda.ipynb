```python
import pandas as pd
import numpy as np

df = pd.read_csv('./sample_data/Most-Recent-Cohorts-Institution.csv', low_memory=False)
df.shape
```

```python
threshold = len(df) * 0.8
df_clean = df.dropna(thresh=threshold, axis=1)
df_clean.shape
```

```python
print(len(list(df_clean.columns[df_clean.dtypes.values == np.dtype('float64')])))
print(len(list(df_clean.columns[df_clean.dtypes.values == np.dtype('int64')])))
print(len(list(df_clean.columns[df_clean.dtypes.values == np.dtype('object')])))
```

```python
df_clean.replace('PrivacySuppressed', np.nan)
```

```python
df_clean.select_dtypes([np.number]).shape
```

```python
df_clean.isna().sum()
```

```python
# trying to remove the 2583 rows which have NaN values
df_clean_no_nan = df_clean.dropna(axis=0)
df_clean_no_nan.shape
```
